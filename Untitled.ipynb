```python
import json

# Replace 'your_file.json' with the path to your JSON file
file_path = 'public/foodmapasia.json'

with open(file_path, 'r') as file:
    data = json.load(file)

# Print the data to verify it's been loaded correctly
print(data)
```

```python
data
```

```python
for prd in data: 
    prd["price"] = str(prd["price"])
    prd["originalPrice"] = str(prd["originalPrice"]) 
                       
```

```python
data
```

```python
output_file = 'foodmap_string_price.json'

with open(output_file, 'w') as file:
    json.dump(data, file)

print(f"Data has been saved to {output_file}")
```
